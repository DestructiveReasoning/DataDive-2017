# Still better than flipping a coin
Memoirs of deriving property prices from AirBnB data 

## Abstract
This project attempts to shade some light to answer the question of wethether or not AirBnB listings can be used as a reliable metric to estimate real estate property values in the Seattle metropolitan area. Motivation for this quesiton comes from the rich amount of features of the property that are listed for each hosts house as well as the fact that the other datasets looked a bit bland.

Build the clean dataframe from the most relevant attributes

```python
pd.options.mode.chained_assignment = None
data = listings[['price', 'review_scores_location', 'monthly_price', 'weekly_price', 'square_feet', 'property_type', 'bedrooms','bathrooms','accommodates', 'zipcode', 'host_neighbourhood', 'review_scores_rating', 'bed_type', 'beds', 'number_of_reviews', 'longitude', 'latitude', 'security_deposit', 'cleaning_fee', 'guests_included', 'availability_365', 'transit', 'reviews_per_month','host_listings_count']]

def nan_to_zero(x):
    if np.isnan(x): return 0
    return x

def parse_bed_type(bed_type):
    if bed_type == 'Real Bed': return 0
    elif bed_type == 'Futon': return 3
    elif bed_type == 'Pull-out Sofa': return 4
    elif bed_type == 'Couch': return 9
    elif bed_type == 'Airbed': return 6
    else: return 1


data['bed_type'] = data.bed_type.apply(parse_bed_type)
data.security_deposit = data.security_deposit.apply(nan_to_zero)
data.cleaning_fee = data.cleaning_fee.apply(nan_to_zero)
data.reviews_per_month = data.cleaning_fee.apply(nan_to_zero)

transformed_transit = data.transit
busCount=np.array([0]*len(data.index))
parkingCount=np.array([0]*len(data.index))
taxiCount=np.array([0]*len(data.index))
walkCount=np.array([0]*len(data.index))
bikeCount=np.array([0]*len(data.index))
emptyCount=np.array([0]*len(data.index))

i = 0

# Count the occurences of transit types specified in the transit column
for entry in transformed_transit:
    #if it's a float, it's actually NaN
    if (isinstance(entry, float)):
        emptyCount[i]+=1
        i+=1
        continue
    busCount[i] = len(re.findall(r'bus', entry, re.IGNORECASE))
    parkingCount[i] = len(re.findall(r'parking', entry, re.IGNORECASE))
    taxiCount[i] = len(re.findall(r'uber|taxi|lyft|car2go', entry, re.IGNORECASE))
    walkCount[i] = len(re.findall(r'walk', entry, re.IGNORECASE))
    bikeCount[i] = len(re.findall(r'bike|biking', entry, re.IGNORECASE))
    i+=1


transit_types = ['bus', 'parking', 'taxi', 'walk', 'bike', 'other_transit']

data['bus'] = busCount
data['parking'] = parkingCount
data['taxi'] = taxiCount
data['walk'] = walkCount
data['bike'] = bikeCount
#any transit not in the above columns
data['other_transit'] = emptyCount

data = pd.merge(data, prop_assessments[['zipcode','bedrooms', 'bathrooms', 'price']], how='inner', on=['zipcode','bedrooms','bathrooms'])
data.rename(columns= {'price_x':'daily_price_airbnb', 'monthly_price':'monthly_price_airbnb', 'weekly_price':'weekly_price_airbnb', 'price_y':'housing_price'}, inplace=True)
# average house price by bathrooms, bedrooms, zipcode
data = data.join(data.groupby(['zipcode','bedrooms', 'bathrooms']).housing_price.mean(), on=['zipcode','bedrooms', 'bathrooms'], rsuffix='_avg')

print('\nSanitized Data:')
data.head()
```

We normalize the values in the dataset by computing their Z-Score, defined as follows:

\begin{equation}
\displaystyle Z_c(x) = \frac{x - \mu_c}{\sigma_c}
\end{equation}

Where $\mu_c, \sigma_c$ are the mean and standard deviation of the values in column $c$, respectively, and $x$ is a value in column $c$. Effectively, this transforms each $x$ into the amount of standard deviations away from the mean. Due to the unlikeliness of being more than approximately 3 standard deviations from the mean, the values become *normalized* to within $\approx\pm 3$, and therefore compare in magnitude to values in other columns.

```python
norm_features = (features - features.mean())/features.std()

# tested with ratios in the range(0.25, 0.1, 0.35) 
X_train, X_test, y_train, y_test = train_test_split(norm_features, data['housing_price_avg'], test_size=0.3)
```

## Making Predictions with the Method of Least Squares

Firstly, we implemented a simple least squares algorithm to analytically determine a relationship between AirBnB data with house prices in a linear model. Linear regression performs the following computation:

\begin{equation}
\mathbf{w} = (\mathbf{X}^T\mathbf{X})^{-1}\mathbf{X}^T\mathbf{y}
\end{equation}

Here, $\mathbf{X}$ is the matrix of independent variables in the training data, and $\mathbf{y}$ is the vector of dependent values in the $\mathbf{y}$. The computation returns a vector $\mathbf{w}$ of weights that are used to predict a house price given AirBnB data. For a vector $\mathbf{x}$ of AirBnB data, the predicted value of house price is computed as $\mathbf{w}^T\mathbf{x}$.

## Learning with Ridge Regression

`sklearn` provides an API for performing Ridge Regression, a form of Linear Regression, that we will optimize a *cost function* that provides a means of predicting house prices based on AirBnB data. The cost function $J$ is a function that takes on greater values for predictions that differ more from expected values ($\mathbf{y}$ values in the function shown below), and is calculated as follows:

\begin{equation}
J = \min_{\mathbf{w}} ||\mathbf{X}\mathbf{w} - \mathbf{y}||_2^2 + \alpha ||\mathbf{w}||_2^2
\end{equation}

Here, $\mathbf{w}$ is a set of weight parameters describing the relative weight of each feature, $\mathbf{X}$ is a matrix of the independent variables of the training data, and \mathbf{y} is the vector of dependent variables of the training data. The $\alpha$ parameter imposes a penalty on high-valued weights, thereby encouraging smaller values in $\mathbf{w}$. This reduces the odds of *over-fitting*, or molding a prediction function too strongly on the training data, which would cause less-than-optimal predictions on separate test data.

### Applying Ridge Regression

```python
from sklearn import linear_model

ns = []
x = range(1,100)
for i in x:
    reg =linear_model.Ridge(alpha = i)
    reg.fit(X_train,y_train)
    ns.append(reg.score(X_test,real_y))
    
fig, arr = plt.subplots(figsize=(20,10))
arr.scatter(x,ns)
plt.title('Model score for different alpha\'s')
plt.xlabel("alpha")
plt.ylabel("Score")

print('Best Alpha: ' + str(max(ns)))
```

## Learning with LASSO Regression

The LASSO Regression is fairly similar to the Ridge Regression seen above, but optimizes a slightly different cost function $\tilde{J}$:

\begin{equation}
\tilde{J} = \min_{\mathbf{w}}\frac{1}{2n}||\mathbf{X}\mathbf{w} -\mathbf{y}||_2^2 + \alpha ||\mathbf{w}||_1
\end{equation}

The prevalent differences here are the scaling by $\frac{1}{2n}$ of the squared error in prediction, where $n$ is the sample size of the training data, and the regularization term is linear in each $\omega\in\mathbf{w}$ rather than quadratic. This effectively reduces the influence of the weight parameters, and scales the influence of the squared error by a constant. This will provide us with information regarding the effects of regularization.

### Applying Lasso Regression

```python
from sklearn import linear_model

ns = []
x = range(1,100)
for i in x:
    reg =linear_model.Lasso(alpha = i, normalize=True)
    reg.fit(X_train,y_train)
    ns.append(reg.score(X_test,real_y))
 
fig, arr = plt.subplots(figsize=(15,12))
arr.scatter(x,ns)
plt.title('Model score for different alpha\'s')
plt.xlabel("alpha")
plt.ylabel("Score")
```

 # Analysis
 
Now that we have seen some examples of the accuracy of our predictions according to the various models that were used, we may reason about what caused success or failure in the predictions.

Firstly, consider the root mean squared error heatmap. In this diagram, the cells with darker colours correspond to zipcodes for which our predictor has a larger aver

Allow us to be frank: these results are not good, per se. That being said, some intuition for our data set may still be obtained, *quod me non interficit, fortiorem me reddit*.

Firstly, note that the general consensus among the linear regressions was that the prediction was more accurate for smaller values of $\alpha$. This implies that regularization harms the prediction. If the prediction was over-fitting, regularization would not always hinder the prediction, so this is an indication that the prediction was severly under-fitted.

Due to the abysmal $R^2$ values in all predictions, it can be concluded that there is little correlation between the data in an AirBnB listing and the value of a house with similar features in the Seattle Metropolitain Area. However, due to the under-fittedness of our predictions, we assumed that perhaps the variance of our data was too high,so we thought it appropriate to try to predict average house values by zipcode. First, the $k$ nearest neighbours method was tested, followed by LASSO regression.

```python
ns = []
x = range(1,10)
for i in x:
    reg = KNeighborsRegressor(n_neighbors=i)
    reg.fit(X_train, y_train)
    ns.append(reg.score(X_test,y_test))
    print('k: {0} -> Score: {1:.5f}'.format(i,ns[i-1]))

fig, arr = plt.subplots()
arr.scatter(x,ns)
plt.title('Model score for different k\'s')
plt.xlabel("k")
plt.ylabel("Score")

ns = []
x = range(1,100)
for i in x:
    reg =linear_model.Lasso(alpha = i, normalize=True)
    reg.fit(X_train,y_train)
    ns.append(reg.score(X_test,y_test))
 
fig, arr = plt.subplots()
arr.scatter(x,ns)
plt.title('Model score for different alpha\'s')
plt.xlabel("alpha")
plt.ylabel("Score")
```

Yippee!

Clearly, predictions were made with considerably more accuracy for the average value of a house in a given region based on AirBnB listings in that region. In particular, the $k$ nearest neighbours method had a prediction function with an $R^2$ of 0.97, which is near perfect.

 # Conlusion
 In retrospect it is pretty clear that property price has very low correlation with AirBnB listings.
 The heavy weight attributed to features derived from the size of the house (i.e. bedrooms, bathrooms, maximum guests) was shown to have probalby no implications on the value of the house: a big house can be found in a cheap neighbourhood. 
 
 With respect to the data, it is unfortunate that we could not find a reliable mapping between the __zipcodes__ and __neighbourhoods__ attributes. Data for __dialy_airbnb_price__ had a greater variance when plotted against __neighbourhoods__ than to __zipcodes__.
 
 Even though we applied preprocessing heuristics such as normalization and test-train splitting, there is a titanic amount of improvement on our part in the field of data analysis.
 
However, based on the accuracy to which the learning methods were able to predict average house prices by zipcode based on AirBnB listings, it can be concluded that the failure to estimate house prices in general was due to the lack of the  predictive power of the features that were used. Unfortunately, we found no data that made for better predictors in the data set given. At this point, we may conclude that either we did not have the correct data to make such predictions, or perhaps there is no way to predict more precise house prices based on AirBnB data alone.
